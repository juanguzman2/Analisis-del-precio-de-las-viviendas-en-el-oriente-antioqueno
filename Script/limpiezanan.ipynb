```python

# Agrega una nueva columna al DataFrame dfdata con valores vacíos
dfdata['Nombre Municipio'] = pd.Series(['']*len(dfdata))

# Itera sobre cada fila del DataFrame dfdata y busca una coincidencia entre las palabras en 'titulo' y 'Nombre Municipio' en el DataFrame dfubi.
for i in range(len(dfdata)):
    for word in dfdata.iloc[i]['titulo'].split():
        if word.upper() in dfubi['Nombre Municipio'].str.upper().values:
            dfdata.at[i, 'Nombre Municipio'] = word.upper()
```
